```python
omega0 = 1
gamma1 = 0.1
gamma2 = -0.1
a0 = 1
```

```python
t_tab = np.linspace(0,50,200)
```

```python
a1 = a0 * np.exp(-1j* omega0* t_tab) * np.exp(gamma1*t_tab)
a2 = a0 * np.exp(-1j* omega0* t_tab) * np.exp(gamma2*t_tab)
```
